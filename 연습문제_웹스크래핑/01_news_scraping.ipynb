### 1. 뉴스 제목 추출하기

```python
url = 'https://news.daum.net/economy'
print(url)

# 요청 헤더 설정 : 브라우저 정보 ( 사람처럼 보이게 하기 위함 )
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
}

# requests 의 get() 함수 호출하기 
res = requests.get(url, headers=req_header)


print(type(res))
print(res.status_code)
# 응답(response)이 OK 이면
# 응답 (response)에서 text 추출
# BeautifulSoup 객체 생성  

res.encoding = 'utf-8'

if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')
    a_tags = soup.select('ul.list_newsheadline2 li')
    print(type(a_tags), len(a_tags))
    for li_tag in a_tags:
        a_tag = li_tag.find('a')
        link = a_tag['href']	
            
        strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
        title = strong_tag.text 
        print(link)
        print(title)

```

```python
import requests
from bs4 import BeautifulSoup

section_dict = {'기후/환경':'climate','사회':'society','경제':'economy','정치':'politics','국제':'world','문화':'culture','생활':'life','IT/과학':'tech','인물':'people'}

def print_news(section_name):
    section = section_dict.get(section_name)
    if not section:
        print(f"Invalid section name: {section_name}")
        return
    url = f'https://news.daum.net/{section}'
    print(f'======> {url} {section_name} 뉴스 <======')
    req_header = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
    }
    res = requests.get(url, headers=req_header)
    
    res.encoding = 'utf-8'

    if res.ok:
        soup = BeautifulSoup(res.text, 'html.parser')
        a_tags = soup.select('ul.list_newsheadline2 li')
        for li_tag in a_tags:
            a_tag = li_tag.find('a')
            link = a_tag['href']	
                
            strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
            title = strong_tag.text 
            print(link)
            print(title)
    else:
        print(f'Error Code = {res.status_code}')
```

```python
print_news('경제')
```

```python
print_news('사회')
```
